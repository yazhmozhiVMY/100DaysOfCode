# Improved version of day 5 using sliding window

```python
def longestUniqueSubstr(s):
    if len(s) == 0:
        return 0
    if len(s) == 1:
        return 1
    maxLength = 0
    visited = [False] * 256
    # l and r pointers for the current window
    left = 0
    right = 0
    while right < len(s):
        # inner while to move left until we exclude the repeated character from the current window
        while visited[ord(s[right])] == True:
            visited[ord(s[left])] = False
            left += 1

        visited[ord(s[right])] = True
        # The length of the current window (right - left + 1) is calculated and answer is updated accordingly.
        maxLength = max(maxLength, right - left + 1)
        right += 1

    return maxLength
```

```python
longestUniqueSubstr("bbbbb")
```

```python
longestUniqueSubstr("abcbadbd")
```

## Space = O(1) assuming character set is constant. We can better say O(Character-Set-Size)
## Time = O(n), since each character is processed by the left and right pointers exactly once.
