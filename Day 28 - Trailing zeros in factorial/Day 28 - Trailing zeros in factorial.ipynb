```python
def countTrailingZerosFactorial(num):
    count = 0
    while num:
        num//=5
        count+=num
    return count
```

```python
countTrailingZerosFactorial(3) 3! = 6
```

```python
countTrailingZerosFactorial(5) # 5! = 120
```

```python
countTrailingZerosFactorial(10) # 10! = 3628800
```

```python
countTrailingZerosFactorial(100) #100! has 24 trailing zeros 
```
