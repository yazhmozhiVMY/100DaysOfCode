#### Given a string containing of ‘0’, ‘1’ and ‘?’ wildcard characters, generate all binary strings that can be formed by replacing each wildcard character by ‘0’ or ‘1’. 

#### Example 1
Input: 
* 10?

Output: 
* 100 
* 101

#### Example 2
Input: s = “1??0?101”

Output: 
* 10000101
* 10001101
* 10100101
* 10101101
* 11000101
* 11001101
* 11100101
* 11101101

#### Your Task:
You don't need to read or print anything. Your task is to complete the function generate_binary_string() which takes the given string as input parameter and returns a vector of strings containing all the possible strings that can be formed.

#### Note : Strings should be printed in lexicographically increasing order.

Expected Time complexity: O(2n)
Expected Space complexity: O(n*2n)

#### Constraints:
1 ≤ length of string ≤ 30
Note: Number of '?' in any string does not exceed 15.

</div>

# Solution 1 - Recursion 

```python
def gen(s,pos):
    if pos == len(s):
        print(''.join(s))
        return
    if s[pos] == '?':
        s[pos]='0'
        gen(s,pos+1)
        s[pos]='1'
        gen(s,pos+1)
        s[pos]='?'
    else:
        gen(s,pos+1)
```

```python
s = "1?1?"

gen(list(s),0)
```

```python
s = "1??0?101"

gen(list(s),0)
```

<div class = "alert alert-block alert-success">

#### Time = O(2^N), where N is the length of the given string and there are 2 possibilities.
#### Space = O(N^2), as a copy of the string is created in every recursive call.
