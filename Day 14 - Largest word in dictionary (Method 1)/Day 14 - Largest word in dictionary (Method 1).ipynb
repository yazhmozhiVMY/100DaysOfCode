```python
class Solution:
    def isSubString(self, str1, str2):
        m = len(str1)
        n = len(str2)
        i = j = 0
        while(i<n and j<m):
            if str1[j] == str2[i]:
                j+=1
            i+=1
        return j==m
    def findLongest(self, dictionary, word):
        maxLen = 0
        result = ""
        sortedDict = sorted(dictionary)
        for w in sortedDict:
            if((len(w) > maxLen) and (self.isSubString(w,word))):
                maxLen = len(w)
                result = w
        return [result,maxLen]
                
```

```python
s = Solution()
s.findLongest({"ale", "apple", "monkey", "plea"},"abpcplea")
```

```python
s1 = Solution()
s1.findLongest({"a", "b", "c"},"abpcplea")
```

```python
s.findLongest({"a", "b", "c"},"zqw")
```

#### Dictionaries do not preserve order. So we need to sort the keys as we need the smallest lexicographical order

<div class="alert alert-block alert-success">

### Time complexity = O(N*(K+n)) 
Here N is the length of dictionary and n is the length of given string ‘str’ and K – maximum length of words in the dictionary.
### Space complexity = O(1)
