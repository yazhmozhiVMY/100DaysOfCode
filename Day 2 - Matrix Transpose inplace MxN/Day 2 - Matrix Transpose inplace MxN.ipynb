# Finding transpose of a mxn matrix inplace algorithm (takes up auxillary mem)

1. Read this first https://www.geeksforgeeks.org/inplace-m-x-n-size-matrix-transpose/
2. Take a close look at this image to understand how nr = oc and nc = or



```python
def MatrixInplaceTranspose( A,  r,  c): 
    size = r * c - 1; 
      
    b = 1; 
      
    b |= (1 << size); 
  
    i = 1; 
    while (i < size): 
      
        cycleBegin = i; 
        t = A[i]; 
        while True: 
            next1 = (i*r)%size; 
            temp = A[next1] 
            A[next1] = t 
            t = temp 
            b |= (1 << i)  
            i = next1; 
            if i == cycleBegin: 
                break
          
        
        i = 1 
        while i < size and ( (b & (1 << i)) != 0): 
            i += 1
          
        print() 
    return A 
```

```python
def Pr2DArray( A,  nr,  nc): 
      
    for r in range(nr):  
        for c in range(nc): 
  
            print('{0: >4}'.format(str(A[r * nc + c])), end = "") 
        print() 
    print() 
```

```python
r = 3
c = 2 
size = r*c 
A = ['a','b','c','d','e','f'] # well this is a 1D list being visualized as a 2D arr :)
```

```python
Pr2DArray(A, r, c)
```

```python
A = MatrixInplaceTranspose(A, r, c); 
```
