### Problem statement


```python
def fun(n,bCnt,cCnt):
    #base cases
    if bCnt<0 or cCnt<0: return 0
    if n==0: return 1
    if bCnt==0 and cCnt==0: return 1
    
    res = fun(n-1,bCnt,cCnt) # choosing a
    res+= fun(n-1,bCnt-1,cCnt) # choosing b
    res+= fun(n-1,bCnt,cCnt-1) # choosing c\
    
    return res
```

```python
fun(1,1,2)
```

```python
fun(3,1,2)
```

```python
fun(4,1,2)
```

```python
fun(3,0,0)
```

### Time Complexity: O(3^N)
### Auxiliary Space: O(1)
