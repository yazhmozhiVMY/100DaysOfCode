```python
arr = [0,-1,2,-3,1]
```

```python
n=len(arr)
for i in range(n):
    for j in range(i+1,n-1):
        for k in range(j+1,n):
            print(i,j,k)
```

<div class="alert alert-block alert-info">

#### Naive approach
</div>

```python
def findZeroSumTriplets(inputArr):
    n=len(inputArr)
    results = []
    for i in range(n):
        for j in range(i+1,n-1):
            for k in range(j+1,n):
                if inputArr[i]+inputArr[j]+inputArr[k]==0:
                    triplet = sorted([inputArr[i],inputArr[j],inputArr[k]])
                    if triplet not in results:
                        results.append(triplet)
    return results
```

```python
inputArr1 = [0,-1,2,-3,1]
findZeroSumTriplets(inputArr1)
```

```python
inputArr2 = [-1,0,1,2,-1,-4]
findZeroSumTriplets(inputArr2)
```

```python
inputArr3 = [0,1,4]
findZeroSumTriplets(inputArr3)
```

```python
inputArr4 = [0,0,0]
findZeroSumTriplets(inputArr4)
```

<div class="alert alert-block alert-success">
    
## Time Complexity: 
O(n^3), with n being the number of elements in the array. This comes from the three nested loops.
## Space Complexity: 
Minimal (ignoring the output space), but the space used increases with the number of valid triplets found.

</div>

#### Optimized approach 
</div>

```python
def findTripletsOptimized(arr):
    arr.sort()
    triplets = []
    for i in range(len(arr) - 2):
        if i > 0 and arr[i] == arr[i - 1]:
            continue
        l, r = i + 1, len(arr) - 1
        while l < r:
            sum = arr[i] + arr[l] + arr[r]
            if sum == 0:
                triplets.append([arr[i], arr[l], arr[r]])
                while l < r and arr[l] == arr[l + 1]:
                    l += 1
                while l < r and arr[r] == arr[r - 1]:
                    r -= 1
                l += 1
                r -= 1
            elif sum < 0:
                l += 1
            else:
                r -= 1
    return triplets
```

#### Time Complexity: 
The sorting operation takes O(nlogn), and the search process takes O(n^2). Thus, the overall complexity is O(n^2), which is a significant improvement over the brute force method.
#### Space Complexity: 
O(1) for the working space (excluding the space needed for storing the output).
