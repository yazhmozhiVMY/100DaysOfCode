```python
class Solution(object):
    def twoSum(self, nums, target):
        """
        :type nums: List[int]
        :type target: int
        :rtype: List[int]
        """
        numIndexDict={}
        for index,num in enumerate(nums):
            pair=target-num
            if pair in numIndexDict:
                return [numIndexDict[pair],index]
            numIndexDict[num]=index
        else:
            return []
```

```python
s1 = Solution()
```

```python
s1.twoSum([2,7,11,15],9)
```

```python
s2 = Solution()
s2.twoSum([3,2,4],6)
```

```python
s3 = Solution()
s2.twoSum([3,3],6)
```
