#### 1. Building a counter 

```python
from collections import Counter
```

```python
inputArr = [1,1,1,2,2,3]
```

```python
count = Counter(inputArr)
```

```python
count[3]
```

```python
count
```

#### Internally, Counter is just a subclass of dict.
1. Time: O(n)
2. Space: O(n)

```python
count.items() # this generates a list of tuples (value, frequency)
```

#### 2. min heap construction using heapq and retrieving the top k freqs

```python
k=2
import heapq
heapq.nlargest(k,count.items())
```

#### We need top k freqs (values) not keys

The key lambda x: x[1] tells Python to compare by frequency.

```python
k=2
import heapq
heapq.nlargest(k,count.items(), lambda x:x[1])
```

```python
resultList = [item for item,freq in heapq.nlargest(k,count.items(), lambda x:x[1])]
```

```python
resultList
```

Time Complexity:
O(n) to build the counter.

O(n log k) for heapq.nlargest.

Total: O(n log k)  better than full sort.

Space Complexity:
O(n) for the counter and heap.
