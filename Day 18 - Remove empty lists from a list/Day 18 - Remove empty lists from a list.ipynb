### Method 1 (Brute force)
</div>

```python
inputList = [[1, 2], [], [3, 4], [], [5,6,7]]
result =[]
for i in inputList:
    if i:
        result.append(i)
print(result)
```

<div class = "alert alert-block alert-success">

#### Time complexity is O(n) n is the number of items in the inputList
#### Space complexity is O(n) if all items are non-empty
</div>

<div class = "alert alert-block alert-info">

### Method 2 (List Comprehension)
</div>

```python
outputList = [item for item in inputList if item]
```

```python
outputList
```

<div class = "alert alert-block alert-success">

#### Time complexity is O(n) n is the number of items in the inputList
#### Space complexity is O(n) if all items are non-empty
</div>

### Method 3 (Lambda and filter)
</div>

```python
fun = lambda b:b
```

```python
a = [[1, 2], [], [3, 4], [], [5]]
```

```python
fun(a)
```

```python
list(filter(lambda b:b, a))
```

```python
result_List = list(filter(lambda b:b, a))
print(result_List)
```

<div class = "alert alert-block alert-success">

#### There is a shorthand notation for the above filter and lambda functions as follows
</div>

```python
list(filter(None, a))
```
