# Method 1 (Brute force)
</div>


```python
inputArr1 = [10,7,6,1]
inputArr2 = [1,3,56,80]
inputArr3 = [1,13,15,2,8]
```

```python
a = [1,10,5,2,8]
a.sort()
print(a)
```

```python
a = [1,10,5,2,8]
a.sort(reverse=True)
print(a)
```

```python
def checkMonotonic1(inputArr):
    x,y=[],[]
    x.extend(inputArr)
    y.extend(inputArr)
    x.sort()
    y.sort(reverse=True)
    return x==inputArr or y==inputArr
```

```python
print(checkMonotonic1(inputArr1))
print(checkMonotonic1(inputArr2))
print(checkMonotonic1(inputArr3))
```

<div class="alert alert-block alert-success">
    
#### Time Complexity: O(N*logN), where N is the length of the array.
#### Auxiliary space: O(N), extra space is required for lists x and y.
</div>

# Method 2 (using a mathematical rule)
</div>


An array is monotonic if and only if it is monotone increasing, or monotone decreasing. Since p <= q and q <= r implies p <= r. So we only need to check adjacent elements to determine if the array is monotone increasing (or decreasing), respectively. We can check each of these properties in one pass.

To check whether an array A is monotone increasing, we’ll check A[i] <= A[i+1] for all i indexing from 0 to len(A)-2. Similarly we can check for monotone decreasing where A[i] >= A[i+1] for all i indexing from 0 to len(A)-2. 

Note: Array with single element can be considered to be both monotonic increasing or decreasing, hence returns “True“. 

```python
arr = [1,2,5,10]
l = [arr[i]<=arr[i+1] for i in range(len(arr)-1)]
print(l)
```

```python
arr = [1,8,5,0]
l = [arr[i]<=arr[i+1] for i in range(len(arr)-1)]
print(l)
```

```python
def checkMonotonic2(inputArr):
    x = all([inputArr[i]<=inputArr[i+1] for i in range(len(inputArr)-1)]) # ascending
    y = all([inputArr[i]>=inputArr[i+1] for i in range(len(inputArr)-1)]) # descending 
    return x or y
```

```python
print(checkMonotonic2(inputArr1))
print(checkMonotonic2(inputArr2))
print(checkMonotonic2(inputArr3))
```

<div class="alert alert-block alert-success">
    
    
#### Time Complexity: O(N), where N is the length of the array.
#### Auxiliary space: O(1) because it is using constant space.
