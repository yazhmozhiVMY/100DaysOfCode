## Method 3 - Dynamic programming (Tabulation)

</div>

```python
def mL(a,n):
    dp = [0]*(n+1) # A table for storing the maximum looted money
    #The table size is n+1 and dp[i] contains the maximum loot from the first i houses
    dp[0] = 0 # Base case 1
    dp[1] = a[0] # Base case 2

    for i in range(2,n+1):
        dp[i] = max(dp[i-1],a[i-1]+dp[i-2]) # dp[i-1] -> we are not picking house i ; a[i-1]+dp[i-2] -> We are picking house i, leaving i-1
    return dp[n]
    
    
```

```python
a=[6,7,1,3,8]
mL(a,len(a))
```

```python
a=[6,7,1,3,8,2,4]
mL(a,len(a))
```

<div class="alert alert-block alert-info">

## Space complexity
O(n) for the table dp
## Time complexity
O(n) - for loop
</div>

## Method 4 - Expected solution (DP with space optimized tabulation)
</div>

```python
def mL_spaceOptimized(a):
    rob=0 # if we rob the current house
    notRob=0 # if we skip the current house
    print(f"Initial values of rob and not Rob are rob={rob} and notRob={notRob}")
    for num in a:
        print(str(num)+"\n=========================\n")
        robCurrent=num+notRob
        notRob=max(notRob,rob)
        rob = robCurrent
        print("Not rob is max(rob,notRob) in the previous iteration and is "+str(notRob))
        print("Rob is looting the current house and is current added to notRob in previous iter "+str(rob))
    return max(rob,notRob)
```

```python
mL_spaceOptimized([6,7,1,3,8])
```

```python
mL_spaceOptimized([6,7,1,3,8,2,4])
```

O(1) no table here
