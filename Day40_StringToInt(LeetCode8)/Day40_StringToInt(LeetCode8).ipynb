```python
class Solution(object):
    def myAtoi(self, s):
        """
        :type s: str
        :rtype: int
        """
        MAX = 2**31-1 #maximum value possible for a 32bit signed integer
        MIN = -2**31 #minimum value possible for a 32-bit signed integer
        index = 0
        length = len(s)
        result = 0
        sign = 1
        # skip leading spaces
        while(index<length and s[index]==' '):
            index+=1
        # checking sign is presentt 
        if (s[index]=="+" or s[index]=="-") and (index<length):
            sign = 1 if s[index]=="+" else -1
            index+=1
        #check numbers
        while(index<length and s[index].isdigit()):
            num =int(s[index])
            #print(f"Num is {num} and result is {result}")
            if(result>MAX//10) or (result==MAX//10 and num>7): #OVERFLOW
                return MAX if sign==1 else MIN
            result=result*10 + num
            index+=1
        return result*sign 
```
