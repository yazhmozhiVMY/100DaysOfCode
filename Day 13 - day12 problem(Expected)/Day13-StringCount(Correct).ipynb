```python
class Solution:
      def countStr(self, n):
          # Case 1: Strings with all 'a'
          strings_all_a = 1
          # Case 2: Strings with at most 1 'b'
          strings_at_most_1_b = n
          # Case 3: Strings with at most 1 'c'
          strings_at_most_1_c = n
          # Case 4: Strings with one 'b' and one 'c'
          strings_one_b_one_c = n * (n - 1)
          # Case 5: Strings with two 'c'
          strings_two_c = (n * (n - 1)) // 2
          # Case 6: Strings with one 'b' and two 'c'
          strings_one_b_two_c = (n * (n - 1) * (n - 2)) // 2
          total_strings = strings_all_a + strings_at_most_1_b + strings_at_most_1_c + strings_one_b_one_c + strings_two_c + strings_one_b_two_c
          return total_strings
      def simplifiedCount(self, n):
          return ((2+(3*n)+(n**3))//2)
          
```

```python
s = Solution()
print(f"length\t\tcountStr\tsimplifiedCount\n")
for i in range(10):
    print(f"{i}\t\t{s.countStr(i)}\t\t{s.simplifiedCount(i)}\n")
```
